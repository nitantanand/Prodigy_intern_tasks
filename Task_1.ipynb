```python
import pandas as pd
import matplotlib.pyplot as plt

dset=pd.read_csv(r"C:\Users\nitan\Desktop\Prodigy\Task 1\API_SP.POP.TOTL_DS2_en_csv_v2_400874\population_v1.csv")
print(dset.columns)
```

```python
print(dset.head(150))
```

```python
countries_plot=['United States', 'Australia', 'China', 'India', 'Ukraine']
year="2021"
filtered_data=dset[dset['Country Name'].isin(countries_plot)]

plt.figure(figsize=(10,5))
plt.bar(filtered_data['Country Name'], filtered_data[year])
plt.xlabel('Country Name')
plt.ylabel(f'Population of Selected COuntries in {year}')
plt.title(f'Population Of Countries in {year}')
plt.xticks(rotation=45, ha='right')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
plt.hist(filtered_data[year], bins=10, edgecolor='black')
plt.xlabel('Population')
plt.ylabel('Frequency')
plt.title(f'Histogram of Population in {year}')
plt.show()
```
